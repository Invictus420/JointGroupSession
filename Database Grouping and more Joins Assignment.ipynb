Question 1

```python
%%sql 
SELECT forename, surname, drivers.driverid, count(*) as wins
FROM drivers 
JOIN results on drivers.driverid = results.driverid
AND results.position = 1
GROUP BY drivers.driverid
ORDER BY wins DESC;
```

Question 2

```python
%%sql 
SELECT drivers.driverref, constructors.name, count(*)
FROM results 
JOIN drivers on drivers.driverid = results.driverid
JOIN constructors on constructors.constructorid = results.constructorid
GROUP BY drivers.driverref, constructors.name
ORDER BY count(*) DESC;
```

Question 3

```python
%%sql 
SELECT drivers.driverref, constructors.name, count(*)
FROM results 
JOIN drivers on drivers.driverid = results.driverid
JOIN constructors on constructors.constructorid = results.constructorid
JOIN status on status.statusid = results.statusid
WHERE status.status = 'Finished'
GROUP BY drivers.driverref, constructors.name
ORDER BY count(*) DESC;
```

Question 4

```python
%%sql 
SELECT driverid, raceid, sum(milliseconds) as total_duration
FROM pitstops 
JOIN drivers USING (driverid)
JOIN races USING (raceid)
GROUP BY driverid, raceid
ORDER BY raceid DESC;
```

```python
%%sql 
CREATE MATERIALIZED VIEW potstops as
SELECT driverid, raceid, sum(milliseconds) as total_duration
FROM pitstops 
JOIN drivers USING (driverid)
JOIN races USING (raceid)
GROUP BY driverid, raceid
ORDER BY raceid DESC;
```

Question 5

```python
%%sql 
SELECT drivers.driverref, constructors.name, count(results.raceid), sum(potstops.total_duration) as dec, count(*)
FROM results 
JOIN drivers on drivers.driverid = results.driverid
JOIN constructors on constructors.constructorid = results.constructorid
JOIN status on status.statusid = results.statusid
RIGHT JOIN potstops ON (potstops.raceid = results.raceid AND potstops.driverid = drivers.driverid)
AND status.status = 'Finished'
GROUP BY drivers.driverref, constructors.name
ORDER BY dec DESC;
```

Question 6.

```python
%%sql 
SELECT drivers.driverref, ps.raceid, constructors.name, sum(ps.total_duration) as duration, count(ps.raceid) as rc, round(sum(ps.total_duration)/count(ps.raceid)) as avg
FROM results 
JOIN drivers on drivers.driverid = results.driverid
JOIN constructors on constructors.constructorid = results.constructorid
JOIN status on status.statusid = results.statusid
JOIN (SELECT driverid, total_duration, raceid FROM potstops) as ps on ps.driverid = drivers.driverid
AND status.status = 'Finished'
GROUP BY drivers.driverref, constructors.name, ps.raceid
ORDER BY avg DESC;
```
